```python
import pandas as pd
df = pd.read_csv("./cilia.csv")
```

```python
df
```

```python
import json

data = {}
for index, row in df.iterrows():
    image_name = row["image_path"].split("\\")[-1]
    caption = row["caption"]
    if row["exposed"] and row["overlaying"]:
        caption += ", both visible on the border of the cell and some other overlaying the body of the cell."
    elif row["exposed"]:
        caption += ", visible on the border of the cell."
    elif row["overlaying"]:
        caption += ", overlaying the body of the cell."
    data[image_name] = caption
```

```python
with open("cilia.json", "w") as json_file:
    json.dump(data, json_file, indent=4)
```
